### Investigación de funciones  

- #### df.dropna:
Elimine los valores que faltan y Quite las etiquetas especificadas de filas o columnas. 

Ejemplo de uso: DataFrame.dropna( eje=0, cómo='cualquiera', thresh=Ninguno, subconjunto=Ninguno, inplace=False) 
 
Eje: {0 o 'índex', 1 o 'columnas'}, predeterminado 0 

Determine si se quitan las filas o columnas que contienen valores que faltan. 

0, o 'índice': coloque filas que contengan valores que faltan. 
1, o 'columnas': coloque columnas que contengan el valor que falta. 

- #### df.astype
Se utiliza para cambiar el tipo de datos de una serie. Cuando el marco de datos se realiza a partir de un archivo csv, las columnas se importan y el tipo de datos se establece automáticamente, lo que muchas veces no es lo que realmente debería tener. Por ejemplo, una columna de salario podría importarse como cadena, pero para realizar operaciones tenemos que convertirla en float. 
astype() se utiliza para realizar dichas conversiones de tipo de datos. 

Sintaxis: DataFrame.astype(dtype, copy=True, errors='raise') 

Parámetros: 
dtype: Tipo de datos para convertir la serie en. (por ejemplo, str, float, int) 
copia: realiza una copia de dataframe/series. 
errores: aumento de errores en la conversión a tipo de datos no válido. Por ejemplo, dict a string. 'raise' planteará el error y 'ignore' pasará sin generar errores.
